```python
def clean_rating(rating_str):
    if pd.isna(rating_str):
        return None
    rating_str = str(rating_str)
    if 'out of' in rating_str:
        rating_str = rating_str.split('out of')[0].strip()
    elif 'ratings' in rating_str:
        return None  # Handle cases like "1,819 ratings" by removing them
    try:
        return float(rating_str)
    except ValueError:
        return None

data['rating'] = data['rating'].apply(clean_rating)
data['rating']
```

```python
#Trail of how the dataset works with basic model

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# Load pre-trained model for semantic search
model = SentenceTransformer('all-MiniLM-L6-v2')

# Standardize category names in dataset: lowercase and strip spaces
data['category'] = data['category'].str.lower().str.strip()

# Extract product descriptions
products = data['product_title'].astype(str).tolist()

# Convert product descriptions into embeddings
embeddings = model.encode(products)

# Define category-accessory mapping (all lowercase)
category_accessory_map = {
    "laptops": ["laptop charger", "laptop bags", "mouse"],
    "smartphone": ["phone charger", "basic cases", "headphone", "screen protector"]
}

# Function for LLM-based search
def search_products(user_query, top_n=20):
    query_embedding = model.encode([user_query])  # Convert query into embedding
    similarities = cosine_similarity(query_embedding, embeddings)  # Compute similarity
    top_indices = np.argsort(similarities[0])[::-1][:top_n]  # Get top N matching products
    return data.iloc[top_indices]  # Return matching products as a sub-dataset

# Function for rule-based filtering
def apply_rule_based_filtering(filtered_dataset):
    # Convert filtered dataset categories to lowercase & strip spaces
    filtered_dataset['category'] = filtered_dataset['category'].str.lower().str.strip()

    unique_categories = filtered_dataset['category'].unique()
    accessory_categories = []

    # Identify relevant accessory categories
    for category in unique_categories:
        if category in category_accessory_map:
            accessory_categories.extend(category_accessory_map[category])

    print("📌 Accessory Categories Identified:", accessory_categories)  # Debugging Step 1

    # Get accessory products from the original dataset
    accessory_products = data[data['category'].isin(accessory_categories)]

    if accessory_products.empty:
        print("⚠️ No accessories found in dataset!")  # Debugging Step 2

    # Ensure at least 5 products per accessory category
    final_accessories = []
    for acc_category in accessory_categories:
        acc_products = accessory_products[accessory_products['category'] == acc_category]
        print(f"🔍 {acc_category} Count: {len(acc_products)}")  # Debugging Step 3

        acc_products = acc_products.sort_values(by=["rating", "review_count", "price"], ascending=[False, False, True]).head(5)  # High rating, high review count, low price

        if not acc_products.empty:
            final_accessories.append(acc_products)

    # Concatenate accessory data
    final_accessory_data = pd.concat(final_accessories) if final_accessories else pd.DataFrame(columns=data.columns)

    # Sort main products by rating & review count (descending)
    filtered_dataset = filtered_dataset.sort_values(by=["rating", "review_count"], ascending=[False, False]).head(10)  # Top 10 main products

    # Append accessories at the bottom of the main recommendations
    final_recommendations = pd.concat([filtered_dataset, final_accessory_data]).reset_index(drop=True)

    return final_recommendations

# Example usage
user_query = input("Enter the query")
filtered_dataset = search_products(user_query)
final_recommendations = apply_rule_based_filtering(filtered_dataset)

# Display final results
print(final_recommendations[['product_title', 'category', 'price', 'rating', 'review_count']])
```
